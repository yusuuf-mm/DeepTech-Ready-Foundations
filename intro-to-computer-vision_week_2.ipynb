## üß© Step 1: Load and Visualize the Corrupted Image

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv_imshow
```

```python
# Load the corrupted image
img_cp = cv2.imread('/content/Corrupted Student Picture.png')
```

```python
# image dimension
img_cp.shape # (height, width, channel)
```

```python
# View and image
cv_imshow(img_cp) # view image with cv_imshow
```

```python
# rgb image
rgb_image = cv2.cvtColor(img_cp, cv2.COLOR_BGR2RGB)
cv_imshow(rgb_image)
```

```python
# Applying contrast & brightness adjustment (optional improvement)
alpha = 1.0  # contrast (1.0 = no change, >1 = stronger contrast)
beta = -9    # brightness (0 = no change, >0 brighter, <0 darker)

corrected = cv2.convertScaleAbs(rgb_image, alpha=alpha, beta=beta)
cv_imshow(corrected)
```

### **Note**

*The image was initially distorted due to incorrect color channel ordering (BGR instead of RGB). After converting it to RGB, the picture returned to its original color. A slight brightness adjustment (alpha = 1.0, beta = ‚Äì9) was applied to improve visibility, while keeping contrast unchanged. This demonstrates how color conversion and basic brightness/contrast tuning can restore image quality using OpenCV.*
